# Preprocessing  
Processing for manually downloaded Morningstar Excel files

```python
from pathlib import Path

data_fp = Path("data")
```

```python
import pandas as pd

df = pd.read_excel(data_fp / "Balance Sheet_Annual_As Originally Reported.xls", engine="xlrd", header=None).T
df.head(3)
```

```python
df.columns=df.iloc[0]
df = df.iloc[1:, :]
df.head(3)
```

```python
df = df.drop(columns=["Fiscal year ends in Dec 31 | USD"])
df.loc[:,df.isna().sum() > 0].head(3)
```

# Annual Cash Flow

```python
def excel_to_df(f_name):
    df = pd.read_excel(data_fp / f_name, engine="xlrd", header=None).T
    df.columns=df.iloc[0]
    df = df.iloc[1:, :]
    df.columns = [c.strip() for c in df.columns]
    df = df.rename(columns={df.columns[0]: "year"})

    return df
```

```python
df = excel_to_df("Cash Flow_Annual_As Originally Reported.xls")
df.head(3)
```

```python
df = df.drop(
    columns = [
        "Other Irregular Income/Loss, Non-Cash Adjustment",
        "Net Movement in Non-Controlling/Minority Interest",
        "Cash Flow Supplemental Section",
        "Fiscal year ends in Dec 31 | USD"
    ]
)

df.head(3)
```

```python

duplicate_missing = [
    "Depreciation, Non-Cash Adjustment", # calculated from dep. and amort. cols
    "Amortization, Non-Cash Adjustment", # calculated from dep. and amort. cols
    "Common Stock Dividends Paid", # duplicated of cash dividends paid
]
df = df.drop(columns = duplicate_missing)
df = df.iloc[:-1,:] # ignore TTM
```

```python
df.to_csv(cleaned_fp / "cash_flow_annual.csv", index=False)
```

```python
duplicate_missing = [
    "Income Statement Supplemental Section",
    "Reported Normalized and Operating Income/Expense Supplemental Section",
    "Discontinued Operations",
    "Fiscal year ends in Dec 31 | USD",
]

df = df.drop(columns=duplicate_missing)
```

```python
df["Reported Normalized Income"] / df["Reported Normalized Operating Profit"]
```

# Quarterly Income Statement

```python
f_name = "Income Statement_Quarterly_As Originally Reported.xls"
df = excel_to_df(f_name)
df = df.iloc[:-1,:]
df["quarter"] = df["year"].str.split().apply(lambda x: x[0][1])
df["year"] = df["year"].str.split().apply(lambda x: x[1])
df.head(3)
```

```python
duplicate_missing = [
    "Income Statement Supplemental Section",
    "Reported Normalized and Operating Income/Expense Supplemental Section",
    "Fiscal year ends in Dec 31 | USD"
]

df = df.drop(columns=duplicate_missing)
```

```python
# only 2 datapoints available. not useful for long term trends
newly_added_features = [
    "Basic EPS",
    "Basic EPS from Continuing Operations",
    "Basic EPS from Discontinued Operations",
    "Diluted EPS",
    "Diluted EPS from Continuing Operations",
    "Diluted EPS from Discontinued Operations",
    "Basic Weighted Average Shares Outstanding",
    "Diluted Weighted Average Shares Outstanding",
    "Reported Normalized Diluted EPS",
    "Total Dividend Per Share",
    "Regular Dividend Per Share Calc",
]

fill_zero = [
    "Discontinued Operations",

]

df[fill_zero] = df[fill_zero].fillna(0)
df = df.drop(columns=newly_added_features)
```

```python
quarterly_income_ratios = (df.groupby("quarter")["Reported Normalized Income"].mean() / df.groupby("quarter")["Reported Normalized Operating Profit"].mean())
quarterly_income_ratios
```

```python
df.loc[df["Reported Normalized Operating Profit"].isna(), "Reported Normalized Operating Profit"] = (
    df.apply(lambda x: x["Reported Normalized Income"] / quarterly_income_ratios[x["quarter"]], axis=1)
)

df.loc[df["Reported Normalized Income"].isna(), "Reported Normalized Income"] = (
    df.apply(lambda x: x["Reported Normalized Operating Profit"] * quarterly_income_ratios[x["quarter"]], axis=1)
)

df[["Reported Normalized Operating Profit", "Reported Normalized Income"]].tail(3)
```

```python
df = df.astype(float)
df.to_csv(cleaned_fp / "income_statement_quarterly.csv", index=False)
```
